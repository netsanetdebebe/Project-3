```python
# Import dependencie 
import pandas as pd 
from pathlib import Path
```

```python
# Files 
general_info = Path('Hospital_General_Information.csv')
locations = Path('hospital_locations.csv')
```

```python
info_df = pd.read_csv(general_info)
info_df.head(10)
```

```python
# Columns 
info_df.columns.values
```

```python
info_df.drop(['index', 'Provider ID', 'Phone Number', 'Location', 'Meets criteria for meaningful use of EHRs', 'Mortality national comparison',
               'Mortality national comparison footnote', 'Safety of care national comparison',
              'Safety of care national comparison footnote', 'Readmission national comparison',
               'Readmission national comparison footnote', 'Patient experience national comparison',
              'Patient experience national comparison footnote','Effectiveness of care national comparison',
               'Effectiveness of care national comparison footnote','Timeliness of care national comparison',
               'Timeliness of care national comparison footnote', 'Efficient use of medical imaging national comparison',
               'Efficient use of medical imaging national comparison footnote','Hospital overall rating footnote'],inplace=True, axis=1)

info_df.head(10)
```

```python
# Missing info
info_df = info_df.dropna(how='any')
```

```python
info_df = info_df.drop_duplicates(subset = "Hospital Name", keep = "first")
info_df
```

```python
# Oregon data
oregon_data = info_df.loc[info_df["State"] == "OR"]
oregon_data
```

```python
# Washington data
washington_data = info_df.loc[info_df["State"] == "WA"]
washington_data
```

```python
# California data
california_data = info_df.loc[info_df["State"] == "CA"]
california_data 
```

```python
#DatFrame
frames = [oregon_data, washington_data, california_data]
westcoast_df = pd.concat(frames)
westcoast_df
```

```python
# Read data
location_df = pd.read_csv(locations)
location_df.head(10)
```

```python
# Columns 
location_df.columns.values
```

```python
# Drop duplicates
location_df = location_df.drop_duplicates(subset = "NAME", keep = "first")
loc_df = location_df.rename(columns={'NAME': 'name', 'ADDRESS': 'address', 'CITY':'city', 'STATE':'state', 'ZIP':'zip', 'TYPE':'type', 'STATUS':'status',
       'POPULATION':'population', 'COUNTY':'county', 'LATITUDE':'latitude', 'LONGITUDE':'longitude', 'OWNER':'owner', 'BEDS':'beds',
       'HELIPAD':'helipad'})
loc_df.head()
```

```python
# loc OR
oregon_loc = loc_df.loc[loc_df["state"] == "OR"]
oregon_loc.head()
```

```python
#LOC WA
washington_loc = loc_df.loc[loc_df["state"] == "WA"]
washington_loc.head()
```

```python
#loc CA
california_loc = loc_df.loc[loc_df["state"] == "CA"]
california_loc.head()
```

```python
#Frames
frames = [oregon_loc, washington_loc, california_loc]
loc_df = pd.concat(frames)
loc_df
```

```python
#Match DatafRames
loc_df = loc_df[loc_df.name.isin(westcoast_df["Hospital Name"])]
loc_df
```

```python
#Westcoast DatafRame

westcoast_df = westcoast_df[westcoast_df["Hospital Name"].isin(loc_df['name'])]
westcoast_df
```

```python
#WC csv
westcoast_df.to_csv('westcoast_info_df.csv',index=False,encoding = 'utf-8')
```

```python
#Save csv
loc_df.to_csv('westcoast_loc_df.csv',index=False,encoding = 'utf-8')
```
